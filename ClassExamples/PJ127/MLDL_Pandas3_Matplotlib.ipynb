```
import pandas as pd
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_no2.csv"
#url_pusa = "https://raw.githubusercontent.com/sarwansingh/Python/master/ClassExamples/data/president_heights.csv"
air_quality = pd.read_csv(url, sep=",")
air_quality 
```

```
air_quality.plot()
```

```
air_quality["station_paris"].plot()
```

```
air_quality.plot.scatter(x="station_london", y ="station_paris",alpha=.5)
```

```
air_quality.plot.box()
```

```
axs = air_quality.plot.area(figsize=(12, 4), subplots=True)
```

```
fig, axs = plt.subplots(figsize=(12, 4));
air_quality.plot.area(ax=axs);
axs.set_ylabel("NO$_2$ concentration");
#fig.savefig("no2_concentrations.png")
```
